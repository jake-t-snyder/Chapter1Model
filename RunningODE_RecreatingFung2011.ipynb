```python
#Timesteps
dt = 1
NUMSTEPS = 10
NUMYEARS = int(NUMSTEPS/dt)
time_points = np.linspace(0, NUMYEARS, NUMSTEPS+1)


#Initial Herbivore Conditions
init_cond = {
    'B0': 0.01,
    'G0': 0.01,
    'S0': 0.01
}


#Scenario
scenario = 'browser_nutr'
#scenario = 'browser_sed'
#scenario = 'grazer_nutr'
#scenario = 'grazer_sed'
#scenario = 'scraper_nutr'
#scenario = 'scraper_sed'
```

```python
if scenario == 'browser_nutr':
    benthos_dictionary = NUTR_benthos
    xparam = np.linspace(0.05, 0.4, 21)
    experiment = experiments_B_NUTR

    for exp in experiment:
        JTS_browser_NUTR_[exp['test_param']] = ODEfunction_BuildSnyder(
            NUMYEARS = NUMYEARS,
            time_points = time_points,
            init_cond = init_cond,
            test_param = exp['test_param'],
            test_param_val = exp['test_param_val'],
            xparam = xparam,
            benthos_dictionary = benthos_dictionary,
            herb_dictionary = exp['herb_dictionary'],
            function = SnyderODE,

    )
        

elif scenario == 'browser_sed':
    benthos_dictionary = SED_benthos
    xparam = np.linspace(0.02, 0.1, 21)
    experiment = experiments_B_SED

    for exp in experiment:
        JTS_browser_SED_[exp['test_param']] = ODEfunction_BuildSnyder(
            NUMYEARS = NUMYEARS,
            time_points = time_points,
            init_cond = init_cond,
            test_param = exp['test_param'],
            test_param_val = exp['test_param_val'],
            xparam = xparam,
            benthos_dictionary = benthos_dictionary,
            herb_dictionary = exp['herb_dictionary'],
            function = SnyderODE,

    )

    
elif scenario == 'grazer_nutr':
    benthos_dictionary = NUTR_benthos
    xparam = np.linspace(0.05, 0.4, 21)
    experiment = experiments_G_NUTR

    for exp in experiment:
        JTS_grazer_NUTR_[exp['test_param']] = ODEfunction_BuildSnyder(
            NUMYEARS = NUMYEARS,
            time_points = time_points,
            init_cond = init_cond,
            test_param = exp['test_param'],
            test_param_val = exp['test_param_val'],
            xparam = xparam,
            benthos_dictionary = benthos_dictionary,
            herb_dictionary = exp['herb_dictionary'],
            function = SnyderODE,

    )


elif scenario == 'grazer_sed':
    benthos_dictionary = SED_benthos
    xparam = np.linspace(0.02, 0.1, 21)
    experiment = experiments_G_SED

    for exp in experiment:
        JTS_grazer_SED_[exp['test_param']] = ODEfunction_BuildSnyder(
            NUMYEARS = NUMYEARS,
            time_points = time_points,
            init_cond = init_cond,
            test_param = exp['test_param'],
            test_param_val = exp['test_param_val'],
            xparam = xparam,
            benthos_dictionary = benthos_dictionary,
            herb_dictionary = exp['herb_dictionary'],
            function = SnyderODE,

    )


elif scenario == 'scraper_nutr':
    benthos_dictionary = NUTR_benthos
    xparam = np.linspace(0.05, 0.4, 21)
    experiment = experiments_S_NUTR

    for exp in experiment:
        JTS_scraper_NUTR_[exp['test_param']] = ODEfunction_BuildSnyder(
            NUMYEARS = NUMYEARS,
            time_points = time_points,
            init_cond = init_cond,
            test_param = exp['test_param'],
            test_param_val = exp['test_param_val'],
            xparam = xparam,
            benthos_dictionary = benthos_dictionary,
            herb_dictionary = exp['herb_dictionary'],
            function = SnyderODE,

    )


elif scenario == 'scraper_sed':
    benthos_dictionary = SED_benthos
    xparam = np.linspace(0.02, 0.1, 21)
    experiment = experiments_S_SED

    for exp in experiment:
        JTS_scraper_SED_[exp['test_param']] = ODEfunction_BuildSnyder(
            NUMYEARS = NUMYEARS,
            time_points = time_points,
            init_cond = init_cond,
            test_param = exp['test_param'],
            test_param_val = exp['test_param_val'],
            xparam = xparam,
            benthos_dictionary = benthos_dictionary,
            herb_dictionary = exp['herb_dictionary'],
            function = SnyderODE,

    )
```

```python
strength = { 
    'xμC': 3,
    'xrM': 3,
    'xrT': 3,
    
    'xαCT': 3,
    'xαMT': 3,

    'xhB': 3,
    'xhG': 3,
    'xhS': 3,

    'xeB': 3,
    'xeG': 3,
    'xeS': 3
}

xμC, xrM, xrT, xαCT, xαMT = strength['xμC'], strength['xrM'], strength['xrT'], strength['xαCT'], strength['xαMT']
xhB, xhG, xhS, xeB, xeG, xeS = strength['xhB'], strength['xhG'], strength['xhS'], strength['xeB'], strength['xeG'], strength['xeS']

poll = load_df(xμC, xrM, xrT, xαCT, xαMT, xhB, xhG, xhS, xeB, xeG, xeS) 

#the poll[#] was just a quick fix. change this later so it's not positional-dependent
#also add this to the larger function. this file should only be setting parameters and such
poll_df = {
    'μC_st': poll[0],
    'rM_st': poll[1],
    'rT_st': poll[2],

    'αCT_st': poll[3],
    'αMT_st': poll[4],

    'hB_st': poll[5],
    'hG_st': poll[6],
    'hS_st': poll[7],

    'eB_st': poll[8],
    'eG_st': poll[9],
    'eS_st': poll[10]
}
```

### Define Ecosystem Component Parameters

```python
benthos = {
    'rC' : 0.2,  #coral intrinsic growth rate
    'αMC': 0.1,  #macroalgae overgrowth over coral scalar
    'μM' : 0.12, #macroalgae mortality rate
    'μT' : 10    #turf mortality rate
}
```

Herbivores

```python
browser = {
    'ηB': 0,       #herbivory scalar
    'sB': 0.12,    #intrinsic growth rate
    'σB': 1,       #carrying capacity scalar
    'θB': 0,       #immigration rate
    'dB': 0.05,    #mortality rate
    'ρB': 1,       #biomass scalar multiplier
    'φB': 0.02     #algae conversion efficiency scalar
}

grazer = {
    'ηG': 0,       #herbivory scalar
    'sG': 0.30,    #intrinsic growth rate
    'σG': 1,       #carrying capacity scalar
    'θG': 0,       #immigration rate
    'dG': 0.05,    #mortality rate
    'ρG': 2.48,    #biomass scalar multiplier
    'φG': 0.02     #algae conversion efficiency scalar
}

scraper = {
    'ηS': 0,       #herbivory scalar
    'sS': 0.37,    #intrinsic growth rate
    'σS': 1,       #carrying capacity scalar
    'θS': 0,       #immigration rate
    'dS': 0.05,    #mortality rate
    'ρS': 3.02,    #biomass scalar multiplier
    'φS': 0.02     #algae conversion efficiency scalar
}

null = {
    'h': 0,       #herbivory rate
    'η': 0,       #herbivory scalar
    's': 0,       #intrinsic growth rate
    'σ': 1,       #carrying capacity scalar
    'θ': 0,       #immigration rate
    'e': 0,       #emigration rate
    'd': 0,       #mortality rate
    'ρ': 0,       #biomass scalar multiplier
    'φ': 0        #algae conversion efficiency scalar
}
```

### Initialize Timesteps, Define Scenario, and Define Initial Herbivore Conditions

1. In these runs, the initial benthic conditions promote AltSS.

```python
#Timesteps
dt = 1
NUMSTEPS = 10
NUMYEARS = int(NUMSTEPS/dt)
time_points = np.linspace(0, NUMYEARS, NUMSTEPS+1)

#Scenario
scenario = "browser"

#Sediment and Nutrient Ranges
sed_value = np.linspace(0, 100, 101)
nutr_value = np.linspace(0, 100, 101)

#Initial Herbivore Conditions
init_cond = {
    'B0': 0.1,
    'G0': 0,
    'S0': 0
}
```

### Subset Sediment Range to Parallelize

```python
run0 = np.linspace(0, 10, 11)
run1 = np.linspace(11, 20, 10)
run2 = np.linspace(21, 30, 10)
run3 = np.linspace(31, 40, 10)
run4 = np.linspace(41, 50, 10)
run5 = np.linspace(51, 60, 10)
run6 = np.linspace(61, 70, 10)
run7 = np.linspace(71, 80, 10)
run8 = np.linspace(81, 90, 10)
run9 = np.linspace(91, 100, 10)
```

### Running ODE

Remember to...
1. Change association strengths
2. Change model parameters
3. Initialize timesteps
4. Define Scenario
5. Define Initial Herbivore Conditions

```python
%%time

runs = np.linspace(0,9,10)

for i in runs: 

    run_d = {
        0: run0,    
        1: run1,    
        2: run2,    
        3: run3,
        4: run4,
        5: run5,
        6: run6,
        7: run7,
        8: run8,
        9: run9,
    }
    
    run_id = run_d[i]
    
    output = ODEfunction(
        scenario,
        run_id,
        dt,
        NUMSTEPS,
        NUMYEARS,
        time_points, 
        sed_value,
        nutr_value,
        init_cond,
        benthos,
        browser,
        grazer,
        scraper,
        null,
        strength,
        poll_df)
    
    output.to_pickle(
        f"equilibria_output{str(round(i))}_011626"
    )
```
